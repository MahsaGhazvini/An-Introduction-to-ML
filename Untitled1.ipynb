```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
INF = 1000000000
NUM_OF_ITERATION = 30
BAD_DIS = 5
n = 2
NUM_OF_DATA = 1000
m = NUM_OF_DATA/10
min_alpha = 0.0
max_alpha = 10.0
max_it = 10000
```

```python
#    MY DATA WITHOUT NOISE
noiseSize = 2

sample = np.arange(NUM_OF_DATA*2).reshape( (2, NUM_OF_DATA) )
sample[1] = np.sqrt(sample)[0]
sample[1] = sample[1] + (np.random.rand(1,NUM_OF_DATA) - .5) * noiseSize*2
sample = sample.T
data = sample
X = data[:, :n-1]
X = np.concatenate((np.ones((1,NUM_OF_DATA)), X.T)).T
Y = np.array( data[:,-1] )

XX = np.copy(X)
YY = np.copy(Y)
    
    
theta = np.ones(n)
plt.plot(XX[:,1], YY, 'yo')
```

```python
x_min = X.min(0)
x_max = X.max(0)
x_mean = X.mean(0)
def feature_scale( feature_num, a):
        return (a - x_mean[feature_num]) / (x_max[feature_num] - x_min[feature_num])
```

```python
def feature_scaling_all(x):
    for j in range(1,n):
        for i in range(0,len(x)):
            x[i][j] = feature_scale(j, x[i][j])
```

```python
def Incode_features_scaling(vec,theta):
    ans = theta[0]
    for i in range(1,n):
        ans += feature_scale(i , vec[i]) * theta[i]
    return ans
```

```python
def linear_regresion(x , y , n , max_a , min_a):
    theta = np.ones(n)
    xT = x.transpose()
    prev_j = -1
    count_it = 0
    exit = False
    while count_it < max_it :
        alpha = (max_a+min_a)/2
        count_it += 1
        h = np.dot(x , theta)
        loss = h - y
        gradient = np.dot(xT , loss)
        theta = theta - alpha * gradient / m
        j = np.sum(loss ** 2) / (2*m)
        if prev_j != -1 :
            if prev_j < j :
                max_a = alpha
                theta = np.ones(n)
                prev_j = -1
                count_it = 0
        prev_j = j;
        if exit == True:
            count_it = max_it
    return theta
```

```python
feature_scaling_all(X)
```

```python
# Normal LinearRegresion Without Noise
theta = linear_regresion(X ,Y ,n , max_alpha , min_alpha)
plt.plot(XX[:,1], YY, 'go')
a = [0,Incode_features_scaling([1,0],theta)]
b = [NUM_OF_DATA-1,Incode_features_scaling([1,(NUM_OF_DATA-1)],theta)]
plt.plot([ a[0], b[0] ], [ a[1], b[1] ], 'k-', lw=1)
h = np.dot(X , theta)
loss = h - Y
```

```python
# MAKE NOISE
for i in range(800, 1000, 5):
    Y[i] = Y[i] - 50
YY = np.copy(Y)
```

```python
# Normal LinearRegresion With Noise
theta = linear_regresion(X ,Y ,n , max_alpha , min_alpha)
plt.plot(XX[:,1], YY, 'ro')
a = [0,Incode_features_scaling([1,0],theta)]
b = [NUM_OF_DATA-1,Incode_features_scaling([1,(NUM_OF_DATA-1)],theta)]
plt.plot([ a[0], b[0] ], [ a[1], b[1] ], 'k-', lw=1)
h = np.dot(X , theta)
loss = h - Y
```

```python
# Linear Regresion Ignoring Outliers
min_bad = INF
best_theta = []
best_index = []

for iterate in range(0,NUM_OF_ITERATION):
    rand_select = random.sample(range(0, NUM_OF_DATA), m)
    x_ = X[ rand_select ]
    y_ = Y[ rand_select ]
    max_alpha = 10.0
    min_alpha = 0.0
    theta = linear_regresion(x_ ,y_ ,n , max_alpha , min_alpha)
    h = np.dot(X , theta)
    loss = h - Y
    bad_data = len(loss[loss < -1*BAD_DIS]) + len(loss[loss > BAD_DIS])
    if min_bad > bad_data:
        min_bad = bad_data
        best_theta = theta
        best_index = rand_select
        
plt.plot(XX[:,1], YY, 'co')
plt.plot(XX[best_index,1],YY[best_index], 'mo')
a = [0,Incode_features_scaling([1,0],best_theta)]
b = [NUM_OF_DATA-1,Incode_features_scaling([1,(NUM_OF_DATA-1)],best_theta)]
plt.plot([ a[0], b[0] ], [ a[1], b[1] ], 'k-', lw=1)
```
